# Data Visualization

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("clean_data.csv")
df.drop(columns = ['package_groups'],inplace=True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = df['platforms'].unique()
fracs = df['platforms'].value_counts()
explode = (0.1, 0.1, 0.1, 0.1)  
fig1, ax1 = plt.subplots()
ax1.pie(x = fracs, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = df['review_score'].unique()
fracs = df['review_score'].value_counts()
#explode = (0.1, 0.1, 0.1, 0.1)  

fig1, ax1 = plt.subplots()
ax1.pie(x = fracs, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
